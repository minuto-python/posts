```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = pd.read_pickle('https://github.com/minuto-python/data/blob/master/regressao-linear/x_rl.pkl?raw=true')
y = pd.read_pickle('https://github.com/minuto-python/data/blob/master/regressao-linear/y_rl.pkl?raw=true')

plt.plot(x, y, 'o')
slope, inter = np.polyfit(x, y, 1)
plt.plot(x, slope * x + inter)

plt.ylabel('Y')
plt.xlabel('X')

plt.show()
```
