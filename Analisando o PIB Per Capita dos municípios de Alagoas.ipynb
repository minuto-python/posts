````python
import geopandas as gpd
import matplotlib.pyplot as plt
````

````python
gdf = gpd.read_file('./dados/pib_al.gpkg')
````

````python
fig, ax = plt.subplots()
fig.suptitle('PIB Per Capita - Alagoas - Censo 2010', fontsize=16)
ax.axis('off')

gdf.plot(ax=ax, column='PIB_percap', legend=True, cmap='Blues', edgecolor='skyblue')

plt.show()
````
